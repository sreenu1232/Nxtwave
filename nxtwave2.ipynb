```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def mobius_strip(u, v, w=1):
    x = (1 + v * np.cos(u/2)) * np.cos(u)
    y = (1 + v * np.cos(u/2)) * np.sin(u)
    z = v * np.sin(u/2)
    return x, y, z
```

```python
u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-0.5, 0.5, 8)
u, v = np.meshgrid(u, v)
```

```python
x, y, z = mobius_strip(u, v)
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b', alpha=0.3)
plt.show()
```

```python
def partial_derivatives(u, v):
    x_u = -np.sin(u) * (1 + v * np.cos(u/2)) - v/2 * np.cos(u) * np.sin(u/2)
    y_u = np.cos(u) * (1 + v * np.cos(u/2)) - v/2 * np.sin(u) * np.sin(u/2)
    z_u = v/2 * np.cos(u/2)
    x_v = np.cos(u/2) * np.cos(u)
    y_v = np.cos(u/2) * np.sin(u)
    z_v = np.sin(u/2)
    # Corrected calculation of E, F, and G using proper Python syntax
    E = x_u**2 + y_u**2 + z_u**2
    F = x_u * x_v + y_u * y_v + z_u * z_v
    G = x_v**2 + y_v**2 + z_v**2
    return E, F, G

def surface_area(u, v):
    E, F, G = partial_derivatives(u, v)
    return np.sqrt(E*G - F**2)

u = np.linspace(0, 2 * np.pi, 100)
v = np.linspace(-0.5, 0.5, 8)
u, v = np.meshgrid(u, v)
area_element = surface_area(u, v)
surface_area_approx = np.sum(area_element) * (u[1,0] - u[0,0]) * (v[1,0] - v[0,0])
print("Approximate Surface Area:", surface_area_approx)
```
